El código

```python
def f1(x):
  return x**2 - 36

#Defino mi intervalo inicial
a = 3
b = 15


print(f"Intervalo inicial: [{a}, {b}]")

#Realizo algunas interaciones del metodo de bisección
for i in range(8): #Aca pongo el nro de iteraciones que quiero que haga
    #Caalculo el punto medio
    c = (a + b) / 2
    #==>Evaluo mi funcion en el punto medio
    fc = f1(c)

    print(f"Iteración {i+1}:")
    print(f" - Punto medio c = {c}")
    print(f" - f(c) = {fc}")

    #Decido el siguiente inetvalo
    if fc == 0:
      print(f"Hemos encontrado la raiz exacta: x = {c}")
      break
    elif f1(a) * fc < 0:
      #La raiz entonces se encuenta en el intervalo [a, c]
      b = c
    else:
      #La raiz entonces va a estar en el intervalo [c, b]
      a = c

    #Imprimo el nuevo intervalo
    print(f"Nuevo intervalo: [{a}, {b}]")
    print()

#Muestro el resultado final despues de las iteraciones
print(f"Despues de 8 iteraciones, la raiz aproximada es: {c}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Definir la variable simbólica y la función
x = sp.symbols('x')
f1_ = x**2 - 16

# Encontrar las raíces de la función
raices_f1 = sp.solve(f1_, x)
print("Raíces de f1(x) =", raices_f1)
print()

def f1(x):
  return x**2 - 16

#Defino mi intervalo inicial
a, b= 3, 15
iterations = 15 #realizo 15 iteraciones
tolerancia = 1e-4
convergio = False #Variable para saber si se alcanzo la tolerancia
#Cuando se cumple la condicion de tolerancia ==> se establece convergio = True
#Lo cual en ese caso significa que se alcanzo la tolerancia y entonces se me imprime
#un mensaje que dice esto, si convergio = False significa que se alcanzo primero el nro.
#de iteraciones_max ==> se va a imprimir el mensaje que indique esto mismo.

#Realizo el metodo de biseccion para las iteraciones que puse
for i in range(iterations):
  c = (a +b) / 2
  #==>Evaluo mi funcion en el punto medio
  fc = f1(c)

  print(f"Iteración {i+1}:")
  print(f" - Punto medio c = {c}")
  print(f" - f(c) = {fc}")
  print()

  #TOLERANCIAAAAAAAAAAA
  #Verificar si f(c) va a estar dentro de mi nivel de tolerancia
  if abs(fc) < tolerancia:
    print(f"Se alcanzó la tolerancia en la iteración {i+1}")
    convergio = True #Alcance mi tolerancia
    break
  elif (f1(a) * f1(c) < 0):
    b = c
  else:
    a = c

  #Imprimo el nuevo intervalo
  print(f"Nuevo intervalo: [{a}, {b}]")
  print()

#Esta version es con algo que busque como hacer, que es que me muestre por separado
#segun cual cosa se alcance primero (tolerancia o iteraciones_max), uso una variable que se
#llama convergio: esta nos dice si la funcion alcanzo el nivel de tolerancia antes de
#agotar la iteraciones.

if convergio:
  print(f"Se alcanzó la tolerancia {tolerancia}, la raiz aproximada es entonces: {c}")
else:
  print(f"Se alcanzó el máximo de iteraciones {iterations}, la raiz aproximada es entocnes: {c}")
```
